| **英文欄位名稱 (Column Name)** | **中文翻譯與用途說明** |

| `""` (Unnamed) | **資料索引**：代表每一列（Row）資料的唯一編號。 |

| `class` | **車輛級距**：車輛的體積與用途分類（如 compact 小型車、suv 休旅車）。 |
