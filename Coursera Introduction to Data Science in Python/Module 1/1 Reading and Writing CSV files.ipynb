#### Step 6. 依照車輛的 `cyl`（汽缸數）進行分類，計算每一種汽缸數對應的平均市區油耗 `cty`。

請根據 `mpg` 資料中的 `cyl` 與 `cty` 欄位，  
計算每一種汽缸數對應的平均市區油耗，  
並將結果依汽缸數由小到大排序後輸出。  

### 解題提示
1. 先找出資料中有哪些不同的 `cyl` 類型

    使用 set() 將 List 轉換為集合，此過程會自動去除重複值

2. 逐一針對每一種 `cyl`，累加該類型所有車輛的 `cty`
3. 再用「總油耗 ÷ 車輛數量」算出平均值，最後整理成清單並排序


cyl	氣缸數量：引擎內部的氣缸總數

cty 城市油耗：在市區道路行駛時的燃油效率

```python
cylinders = set(d['cyl'] for d in mpg)
cylinders
```

對每一種 `cyl`，準備兩個變數：  
`summpg`用來累加 `cty`，`cyltypecount`用來計算該類型共有幾筆資料
當資料列的 `cyl` 符合目前類型時，就進行累加與計數。  

```python
CtyMpgByCyl = []

for c in cylinders:
    summpg = 0
    cyltypecount = 0
    for d in mpg:
        if d['cyl'] == c :
            summpg += float(d['cty'])
            cyltypecount += 1
    CtyMpgByCyl.append((c, summpg/ cyltypecount))
CtyMpgByCyl.sort(key=lambda x: x[0])
CtyMpgByCyl
```

### Step 7.：計算各車輛類別的平均高速油耗
請根據 `mpg` 資料中的 `class` 與 `hwy` 欄位，  
統計每一種車輛類別的平均高速油耗，  
並將結果依平均油耗由小到大排序後輸出。  

`class` 車輛的體積與用途分類（如 compact 小型車、suv 休旅車）。

`hwy` 高速公路油耗：在高速公路行駛時的燃油效率

```python
vehicleclass = set(d['class'] for d in mpg)
vehicleclass
```

### 💡 台灣市場的小補充

- **Subcompact vs. Compact：**
    
    在台灣，我們有時候會把這兩者通稱為「小轎車」。但嚴格來說，像是 **Yaris** 這種長度在 4 公尺左右的算 **Subcompact**；而 **Altis** 這種長度接近 4.6 公尺的標準房車則屬於 **Compact**。
    
- **Crossover (跨界休旅)：**
    
    這是清單上沒提到但在台灣**超熱賣**的分類（例如 **Toyota Corolla Cross**）。它的大小介於 Subcompact 和 SUV 之間，具備休旅車的外型但車身較小，非常適合台灣狹窄的街道。

```python
HwyMpgByClass = []

for t in vehicleclass : # iterate over all the vehicle classes
    summpg = 0
    vclasscount = 0
    for d in mpg:
        if d['class'] ==t:
            summpg += float(d['hwy'])
            vclasscount += 1 # increment the count
    HwyMpgByClass.append((t, summpg/ vclasscount)) # append the tuple ('class', 'avg mpg')

HwyMpgByClass.sort(key = lambda x: x[1])
HwyMpgByClass
```
