##  使用 Python 處理 csv 檔案 ， Code : `%precision 2` ，意義為何?

`%precision 2` 並非 Python 語言的標準語法


這是一個專屬於 **IPython** 與 **Jupyter Notebook** 互動式環境中的「魔術指令」 (Magic Command)。其核心意義在於 **全域設定環境中浮點數 (Floating-point numbers) 的顯示精度為小數點後兩位**

- 在Jupyter等互動式環境中， 以 `%` 開頭的指令被稱為「行魔術指令」(Line Magics)，它們是環境提供給使用者的特殊控制語法，而非 Python 直譯器原生的程式碼

- 此指令僅能在 IPython Console 或 Jupyter Notebook 的儲存格 ( Cell ) 中被解析與執行

- 此指令最關鍵的特性在於 : 它僅改變數值在螢幕上的「顯示方式」，絲毫不影響記憶體中變數的「實際儲存精度」與運算結果

- 假設系統中有一個浮點數變數 x ，其記憶體中的高精度數值為:

# x = 3.1415926535…

當執行 `%precision 2` 後， Jupyter 環境在印出該數值時，會自動進行格式化，顯示為 3.14。 但若將 x 投入後續的數學運算，系統依然會使用底層的3.1415926535… 進行運算，確保資料處理的精準度不受顯示設定影響

### 在 CSV 資料處理中的實際作用

在處理真實世界的資料表時，從 CSV 載入的數值資料經常包含冗長的小數。當資料被讀取或轉換為 NumPy 陣列結構，或者具有多個列 (Row) 與 行 (Column) 的 Pandas DataFrame 時，過多的小數位數會導致終端機或Notebook介面的輸出變得雜亂，降低人類閱讀與初步檢查資料狀態的效率

執行 `%precision 2` 後，後續印出的陣列或數值便會以對齊、簡潔的小數點後兩位呈現，大幅提升多列、多行資料在視覺上的可讀性

( 註 : Pandas 本身亦有獨立的顯示精度控制選項 `pd.options.display.precision` ，但 `%precision` 能廣泛作用於 Python 基礎浮點數、NumPy 陣列等綜合輸出 )

### 程式碼範例 (Use JupyterNote book)

```python
import numpy as np

# 模擬從 CSV 檔案某一行 (Column) 萃取出的數值資料
data_array = np.array([12.345678, 98.765432, 5.555555])

# 1. 預設顯示狀態 (會印出完整精度的浮點數)
print("預設輸出 : ")
data_array
```

```python
# 2. 執行魔術指令，設定顯示精度為小數點後 2 位
%precision 2

# 3. 再次印出同一變數，結果將會改變
print("\n 設定 %precision 2 後的輸出: ")
data_array
#預期顯示結果類似於 : array([12.35, 98.77,  5.56])
```

```python
# 驗證底層資料並未被修改
# 提取第一列，第一行(Column)的元素進行運算
print("\n底層實際數值:", data_array[0])
```
