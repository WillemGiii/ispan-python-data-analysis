# Python Anonymous Function Lambda

`lambda`  函式 (又稱為匿名函式，Anonymous Function) 是一種建構輕量級、即用即棄函式的簡潔語法。

與使用 `def` 關鍵字定義的常規函式不同， `lamda` 函式省略了命名與標準區塊結構，特別適合用於需要將簡單邏輯作為引數傳遞給高階函式 (Higher-order functions) 的場景

### 語法結構

`lambda` 函式的標準語法 :

```python
lambda arguments: expression
```

`arguments` 可以是多個引數 (以逗號分隔)，而 `expression` 必須是一個 **單一的表達式**

Python 會自動計算並回傳該表達式的結果，因此不需要明確撰寫 `return` 關鍵字

#### 舉例

定義函數

$$
f(x, y) = x^2 + y^2
$$

在 Python 中可以使用 `lambda` 進行等價轉換：

```python
f = lambda x, y : x**2 + y**2
result = f(3, 4)
# 運算結果為 25
```

#### 底層機制 與 `def` 的比較

在 Python 的執行環境中，無論是透過 `def` 或 `lambda` 建立的函式，最終都會被編譯為 `<class 'function'>` 的物件。兩者的核心差異在於命名空間與語句限制

| 特性 | `def` 關鍵字宣告 | `lambda`  匿名函式 |
| --- | --- | --- |
| **識別名稱( `__name__`)** | 具備明確的函式名稱 | 預設為 `<lambda>` |
| **邏輯複雜度** | 可包含多行陳述式(Statements)、迴圈與例外處理 | 僅限單一表達式 (Expression)，不允許賦值或迴圈語法 |
| **生命週期** | 通常存在於整個模組或作用域中 | 通常作為暫時性物件，傳遞後即由垃圾回收機制(GC)處理 |

#### 實務應用範例

在資料處理與演算法實作中， `lambda` 最常與 `map()` 、 `filter()` 以及排序函式 `sorted()` 結合使用。

**範例 - 結合高階函式 `map` 與 `filter`**

````python
# 隨機感測器數據 (Celsius)
temperature_data = [ 22.5, 30.1, 15.8, 35.2 ]

# 使用 lambda 將攝氏轉為華氏 : F = C * 9/5 + 32
fahrenheit_data = list(map(lambda c : c * 9/5 +32, temperature_data))

print("華氏溫度:", fahrenheit_data)

# 使用 filter 與 lambda 篩選出異常高溫 ( 大於 30 度 )
highj_temp_alerts = list(filter(lambda x: x> 30 , temperature_data))

print("異常高溫(大於 30 度):", highj_temp_alerts)
````

### 1. `map()` 函式：資料映射與轉換 (Data Transformation)

**定義與作用：** `map(function, iterable)` 的核心職責是「轉換」。它會將傳入的 `function` 規則，逐一應用（Apply）到 `iterable` 中的「每一個」元素上，並將轉換後的結果依序封裝成一個新的迭代器回傳。

#### 範例 : 將定價陣列統一乘以匯率 ( 轉換 ) -- 使用 map() 函式

````python
# 資料 : 商品基礎定價
base_prices = [100, 200, 300]

# 將所有價格乘以 1.05 (情境 -> 匯率上升)
adjusted_prices = list(map(lambda p: p*1.05, base_prices))

print("調整後的價格:", adjusted_prices)
````

---

#### 2. `filter()` 函式：資料篩選與過濾 (Data Filtering)

**定義與作用：**

`filter(function, iterable)` 的核心職責是「篩選」。它會將傳入的 `function` 作為布林判斷條件（Predicate），逐一測試 `iterable` 中的每一個元素。

唯有當 `function(element)` 回傳 `True` 時，該元素才會被保留並放入新的迭代器中；若回傳 `False` 則會被直接捨棄。

#### 範例 : 篩選超過特定閾值的數值 -- 使用 filter() 函式

````python
# 資料 : 系統運算延遲時間 (毫秒)
latency_data = [12, 55, 6, 102, 33]

# 篩選出延遲大於 50 毫秒的異常紀錄
timeout_errors = list(filter(lambda x : x >50, latency_data))

print("大於50毫秒的異常紀錄", timeout_errors)
````

`map()` 與 `filter()` 在資料處理管線（Data Pipeline）中扮演著分工明確的角色：

- 當任務目標是 **「改變資料」**（例如：單位換算、型態轉換、字串清理），且需保留完整資料筆數時，應採用 `map()`。

- 當任務目標是 **「剔除資料」**（例如：排除異常值、尋找特定特徵樣本），從而縮減資料規模時，應採用 `filter()`。
