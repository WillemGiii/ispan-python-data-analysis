假設有兩個集合 A 與 B :

\- A 代表選修「Python 網路爬蟲」的學生名單

\- B 代表選修「資料視覺化」的學生名單

```python
# 範例資料
set_A = {"User_01", "User_02", "User_03"}
set_B = {"User_02", "User_03", "User_04"}
```

#### 1. 聯集(Union)
包含 A 或 B 中的元素

數學定義:

$$A \cup B = \{x \mid x \in A \lor x \in B\}$$

```python
# 使用 | 運算子
union_set = set_A | set_B
union_set
```

#### 2. 交集 (Intersection)
同時存在於 A 或 B 中的元素

數學定義:

$$A \cap B = \{x \mid x \in A \land x \in B\}$$

```python
# 使用 & 運算子
intersection_set = set_A & set_B
intersection_set
```

##### 3 差集 (Difference)

存在於 $A$ 中，但不存在於 $B$ 中的元素。

數學定義：

$$A \setminus B = \{x \mid x \in A \land x \notin B\}$$

```python
# 使用 - 運算子
difference_set = set_A - set_B
difference_set
```

#### 4 對稱差集 (Symmetric Difference)

存在於 $A$ 或 $B$ 中，但不同時存在於兩者中的元素。

數學定義：

$$A \triangle B = (A \setminus B) \cup (B \setminus A)$$

```python
# 使用 ^ 運算子
sym_diff_set = set_A ^ set_B
# 結果: {'User_01', 'User_04'}
sym_diff_set
```
