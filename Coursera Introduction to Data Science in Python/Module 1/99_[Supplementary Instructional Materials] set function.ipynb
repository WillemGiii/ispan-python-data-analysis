#### 集合可以透過大括號 `{}` 或內建函式 `set()` 來建立

```python
# 建立包含不重複元素的集合
valid_course = {'Python基礎', '機器學習','深度學習'}

# 使用 set() 將 List 轉換為集合，此過程會自動去除重複值
raw_student_ids = [101, 102, 101, 103, 104, 102]
unique_student_ids = set(raw_student_ids)

print(f"去重複後的學號:{unique_student_ids}")

# 建立空集合的正確方式
empty_set = set()
```

假設有兩個集合 A 與 B :

\- A 代表選修「Python 網路爬蟲」的學生名單

\- B 代表選修「資料視覺化」的學生名單

```python
# 範例資料
set_A = {"User_01", "User_02", "User_03"}
set_B = {"User_02", "User_03", "User_04"}
```

#### 1. 聯集(Union)
包含 A 或 B 中的元素

數學定義:

$$A \cup B = \{x \mid x \in A \lor x \in B\}$$

```python
# 使用 | 運算子
union_set = set_A | set_B
union_set
```

#### 2. 交集 (Intersection)
同時存在於 A 或 B 中的元素

數學定義:

$$A \cap B = \{x \mid x \in A \land x \in B\}$$

```python
# 使用 & 運算子
intersection_set = set_A & set_B
intersection_set
```
